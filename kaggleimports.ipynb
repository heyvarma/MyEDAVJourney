```python
import pandas as pd

df = pd.read_csv('insurance.csv',nrows = 15)

print(df)



```

```python
agesmoker_df = df[['age', 'smoker']]
print(agesmoker_df)
avg = agesmoker_df.groupby('smoker')['age'].mean()
print(avg)
```

```python
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '60+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

smokeragegroup = df.groupby(['age_group', 'smoker']).size()
print(smokeragegroup)
```

```python
import matplotlib.pyplot as plt
y = df['charges']
x = df['age']
plt.plot(x,y)
```

```python
averagecharges = df.groupby('smoker')['charges'].mean()

print(averagecharges)


df['predictedcharges'] = df['smoker'].map(averagecharges)

print(df[['smoker', 'charges', 'predictedcharges']])
```
